Prediction Functions

```python
def spell_check(sentence):
  has_gibberish = False
  doc = nlp(sentence)
  for token in doc:
    if not token.has_vector:
      has_gibberish = True
  if has_gibberish == True:
    correction = str(TextBlob(sentence).correct())
    doc = nlp(correction)
    for token in doc:
      if not doc.has_vector:
        return "gibberish"
    return correction
  else:
    return sentence

def bag_of_words(sentence, words):
  tokens = nlp(sentence)
  pattern_words = [token.lemma_.lower() for token in tokens]
  bag = [1 if word in pattern_words else 0 for word in words]

  data = np.array(bag)
  data = data.reshape(1, -1)

  return data

def predict(sentence, model, words, classes):
  sentence = spell_check(sentence)
  if sentence == 'gibberish':
    return 'na'
  input = bag_of_words(sentence, words)
  output = model.predict(input)
  if np.max(output) < 0.65:
    return 'na'
  predicted_index = np.argmax(output)
  predicted_class = classes[predicted_index]
  return(predicted_class)

def find_similarity(sentence, list_A, list_B, response_A, response_B):
    doc = nlp(sentence)
    response_A_similarity = []
    response_B_similarity = []
    for item in list_A:
      for token in doc:
        response_A_doc = nlp(item)
        response_A_similarity.append(token.similarity(response_A_doc))
    for item in list_B:
      for token in doc:
        response_B_doc = nlp(item)
        response_B_similarity.append(token.similarity(response_B_doc))
    max_response_A = max(response_A_similarity)
    max_response_B = max(response_B_similarity)
    if (max_response_A < 0.7) and (max_response_B < 0.7):
      return 'na'
    elif max_response_A > max_response_B:
      return response_A
    elif max_response_B > max_response_A:
      return response_B
    else:
      return 'na'
```

```python
def reset():
  print('Reseting...')
  return main()

def main():
  global user_input
  global revenue
  global orders_food_check
  global food_recommendation_check

  orders_food_check = False
  food_recommendation_check = False
  greeting_check = False
  understand_check = False

  while True:
    if understand_check == True:
      print("Can you try saying that again please?")

    if (greeting_check == False) and (understand_check == False):
      print("Hi there! I'm Eatsy, your digital dining companion, ready to guide you on a mouthwatering journey through culinary delights!")
      print("What can I do for you? (Just give me a greeting if you wanna know what I'm capable of)")
    elif (greeting_check == True) and (understand_check == False):
      print("What can I do for you?")

    understand_check = False

    user_input = (input("You: ")).lower()

    if user_input == 'reset':
      return reset()
    elif user_input == 'back':
      return main()

    elif predict(user_input, intents_model, words, classes) == 'greetings':
      print('Here are some of the things I can do for you:')
      print("1. If you want to start an order, I'll get that done for you right away!")
      print("2. Want to book a table? I can help you with reservations too, all you gotta do is ask!")
      print('3. Have any dietery restrictions? Well just ask for a recommendation, and I can help you to the best of my ability!')
      print("4. for customers who have encountered any issues so far, just let me know if you want to write a report, and I'll make sure that you message gets delivered!")
      print("4. And finally, if you ever wanna go back to the previous section, or restart our conversation, simply type 'back' or 'reset' at any point you want.")
      greeting_check = True
      continue

    elif predict(user_input, intents_model, words, classes) == 'Order':
        print('Sure! We will get into it right away.')
        return orders_food()

    elif predict(user_input, intents_model, words, classes) == 'reservation':
        print("Sure, we are ready to take your reservation")
        return table_reservation()

    elif predict(user_input, intents_model, words, classes) == 'Recommendations':
        return food_recommendation()

    elif predict(user_input, intents_model, words, classes) == 'report':
        return issue_report()

    else:
      print("sorry, I did not understand.")
      understand_check = True
      continue

#=================================================================================================================================================================================================================

def orders_food():
  global user_input
  global table_confirmed
  global table_count
  global orders_food_check

  dine_in_list = ['dine in', 'eat in', 'eat inside', 'eat at the restaurant', 'in the restaurant', 'dine', 'in', 'inside', 'dinein', 'dine-in']
  takeaway_list = ['takeaway', 'take away', 'out', 'outside', 'to go', 'togo', 'take it away', 'eat out']

  print('Would you like to Dine-in or Takeaway?')
  user_input = (input("You: ")).lower()
  if user_input == 'back':
    return main()
  elif user_input == 'reset':
    return reset()

  orders_food_response = find_similarity(spell_check(user_input), dine_in_list, takeaway_list, 'Dinein', 'Takeaway')

  if orders_food_response == 'Dinein':
    print('Thank you for your confirmation.')

    check = True
    while check == True:
      print('May we know your table number please? (We have 50 tables at our establishment)')
      user_input = (input("You: ")).lower()
      dine_in_response = spell_check(user_input)
      dine_in_doc = nlp(dine_in_response)

      table_number_check = ['We are at', 'table number', 'number', 'table', 'over at', 'I am at', 'I am in', 'im at', 'im in', 'in at', 'at', 'in']
      table = ''

      if user_input == 'back':
        orders_food_response = ''
        return orders_food()
      elif user_input == 'reset':
        return reset()

      for phrase in table_number_check:
        doc = nlp(phrase)
        if (dine_in_doc.similarity(doc) > 0.5) or (token.like_num for token in dine_in_doc):
          for i, token in enumerate(dine_in_doc):
            if token.tag_ == 'CD':
              if i < len(dine_in_doc) - 1 and dine_in_doc[i+1].tag_ == 'CD': #In case its a two digit number that has been written in text (Eg: Twenty Four)
                table = token.text + " " + dine_in_doc[i+1].text
                break
              elif i < len(dine_in_doc) - 2 and dine_in_doc[i+2].tag_ == 'CD': #In case a two digit number has been seperated by a dash (Eg: Twenty-Four)
                table = token.text + " " + dine_in_doc[i+2].text
                break
              else:
                table = token.text
                break

      if table == '':
        print("Sorry, but we haven't received a valid table number yet.")
        dine_in_response = ''
        continue
      elif table.isnumeric():
        table = int(table)
      elif table.replace(' ', '').isalpha():
        w2n = Word2Num(fuzzy_threshold=60)
        table = w2n.parse(table)
      else:
        print("Sorry, but the number that you have entered is invalid.")
        dine_in_response = ''
        continue

      if (0 < table <= table_count):
        table_confirmed = str(table)
        print('Thank you. We have confirmed that you are at table number: ' + table_confirmed)
        check = False
        print('We are ready to take your order now.')
        orders_food_check = True
        return menu_options()
      else:
        print("Sorry, but the number that you have entered is invalid. We only have 50 tables here.")
        dine_in_response = ''
        continue


  elif orders_food_response == 'Takeaway':
    check = True
    while check == True:
      print("Takeaway has been selected. Do you want to continue?")
      user_input = (input("You: ")).lower()

      if user_input == 'back':
        check = False
        return orders_food()
      elif user_input == 'reset':
        check = False
        return reset()

      takeaway_response = find_similarity(spell_check(user_input), yes_list, no_list, 'yes', 'no')

      if takeaway_response == 'yes':
        print('Ok! You may collect the order from counter number ' + str(random.randint(1, 6)) + ' once it has been prepared.')
        check = False
        print('We are ready to take your order now.')
        orders_food_check = True
        return menu_options()
      elif takeaway_response == 'no':
        check = False
        return orders_food()
      else:
        print('Sorry, I did not understand.')
        continue

  else:
    print('Sorry, I did not understand.')
    return orders_food()

#=================================================================================================================================================================================================================

def table_reservation():
  global today
  global today_formatted

  print("Today's date is " + today_formatted)
  print('Kindly provide the DATE and TIME of your preferred booking')
  print('We are open 7 days a week, for all 24 hours')
  print('(Please note that we ONLY ACCEPT bookings up to 7 days in advance)')
  user_input = (input("You: ")).lower()
  date = user_input

  if user_input == 'back':
    return main()
  elif user_input == 'reset':
    return reset()

  try:
    date = parse(date)
    seven_days_from_today = today + timedelta(days=7)

    if date > seven_days_from_today:
      print("Sorry, but the date you have provided falls outside the booking window of 7 days.")
      return table_reservation()

    elif date < today:
      print("Sorry, but the date you have provided is invalid.")
      return table_reservation()

    else:
      print("The date and time you have entered is: " + date.strftime('%d %B, %Y %I %M %p'))
      print("Are you sure?")
      user_input = (input("You: ")).lower()
      confirmation_response = find_similarity(spell_check(user_input), yes_list, no_list, 'yes', 'no')

      if user_input == 'back':
        return table_reservation()
      elif user_input == 'reset':
        return reset()

      check = False
      while check == False:
        if confirmation_response == 'yes':
          check = True
          print('Confirmation Noted')

          check2 = False
          while check2 == False:
            print('May we know how many people are going to be attending? (10 Max)')
            user_input = (input("You: ")).lower()
            number_of_people = user_input
            number_of_people_doc = nlp(number_of_people)

            for i, token in enumerate(number_of_people_doc):
              if token.tag_ == 'CD':
                if i < len(number_of_people_doc) - 1 and number_of_people_doc[i+1].tag_ == 'CD': #In case its a two digit number that has been written in text (Eg: Twenty Four)
                  people = token.text + " " + number_of_people_doc[i+1].text
                  break
                elif i < len(number_of_people_doc) - 2 and number_of_people_doc[i+2].tag_ == 'CD': #In case a two digit number has been seperated by a dash (Eg: Twenty-Four)
                  people = token.text + " " + number_of_people_doc[i+2].text
                  break
                else:
                  people = token.text
                  break

            if people == '':
              print("Sorry, but we haven't received a valid  number yet.")
              continue
            elif people.isnumeric():
              people = int(people)
            elif people.replace(' ', '').isalpha():
              w2n = Word2Num(fuzzy_threshold=60)
              people = w2n.parse(people)
            else:
              print("Sorry, but the number that you have entered is invalid.")
              continue

            if (0 < people <= 10):
              print('The number of people you requested for is: ' + str(people))
              print('Is that correct?')
              user_input = (input("You: ")).lower()
              confirmation_response2 = find_similarity(spell_check(user_input), yes_list, no_list, 'yes', 'no')

              if user_input == 'back':
                continue
              elif user_input == 'reset':
                return reset()

              if confirmation_response2 == 'yes':
                check2 = True
                print('Thank you for your confirmation.')
                print('Your table has been reserved.')
              elif confirmation_response2 == 'no':
                continue
              else:
                print('Sorry, but the number you have entered is invalid.')
                continue
            else:
              print('Sorry, but the number you have entered is invalid.')
              continue

        elif confirmation_response2 == 'no':
          check = True
          return table_reservation()
        else:
          print("Sorry, I did not understand.")
          continue



  except KeyboardInterrupt:
    sys.exit()

  except Exception as e:
    print("Sorry, but the response provided was invalid")
    return table_reservation()

#=================================================================================================================================================================================================================

def food_recommendation(): #Non-vegetarian, vegetarian, vegan, diabetes, gluten intolerance, lactose intolerance, ketogenic, pescatarian, halal, no INGREDIENT(covers allergies)
  global food_recommendation_check

  print("Let me know about your preferences and dietary restrictions.")
  print("I will let you know what items best suit your needs?")
  user_input = (input("You: ")).lower()

  if user_input == 'reset':
    return reset()

  if user_input == 'back':
    return main()


  food_recommendation_doc = nlp(user_input)

  prefer_similarity = 0
  dont_prefer_similarity = 0
  non_vegetarian_similarity = 0
  vegetarian_similarity = 0
  vegan_similarity = 0
  diabetes_similarity = 0
  keto_similarity = 0
  allergy_similarity = 0

  prefer = False
  dont_prefer = False
  non_vegetarian = False
  vegetarian = False
  vegan = False
  diabetes = False
  keto = False
  allergy = False

  for i, token in enumerate(food_recommendation_doc):
    if token.tag_ == 'VBP' or token.tag_ == 'VB' or token.tag_ == 'RB':
      for j, item in enumerate(food_rec["Prefer"]):
        if token.similarity(nlp(item)) > prefer_similarity:
          prefer_similarity = token.similarity(nlp(item))

    if token.tag_ == 'VBP' or token.tag_ == 'VB' or token.tag_ == 'RB':
      for j, item in enumerate(food_rec["Don't Prefer"]):
        if token.similarity(nlp(item)) > dont_prefer_similarity:
          dont_prefer_similarity = token.similarity(nlp(item))

    for j, item in enumerate(food_rec["Non-Vegetarian"]):
      if (token.similarity(nlp(item)) > non_vegetarian_similarity):
        non_vegetarian_similarity = token.similarity(nlp(item))

    if 'nonveg' in token.text:
      non_vegetarian_similarity = 1

    for j, item in enumerate(food_rec["Vegetarian"]):
      if token.similarity(nlp(item)) > vegetarian_similarity:
        vegetarian_similarity = token.similarity(nlp(item))
        if any('non' in token.text for token in food_recommendation_doc[0:i]):
            non_vegetarian_similarity = 1

    if 'vegan' in token.text:
      vegan_similarity = 1

    for j, item in enumerate(food_rec["Diabetes"]):
      if token.similarity(nlp(item)) > diabetes_similarity:
        diabetes_similarity = token.similarity(nlp(item))

    for j, item in enumerate(food_rec["Keto"]):
      if token.similarity(nlp(item)) > keto_similarity:
        keto_similarity = token.similarity(nlp(item))

    for j, item in enumerate(food_rec["Allergy"]):
      if token.similarity(nlp(item)) > allergy_similarity:
        allergy_similarity = token.similarity(nlp(item))

  if prefer_similarity > 0.9:
    prefer = True
  if dont_prefer_similarity > 0.9:
    dont_prefer = True
  if non_vegetarian_similarity > 0.9:
    non_vegetarian = True
  if vegetarian_similarity > 0.9:
    vegetarian = True
  if vegan_similarity > 0.9:
    vegan = True
  if diabetes_similarity > 0.9:
    diabetes = True
  if keto_similarity > 0.9:
    keto = True
  if allergy_similarity > 0.9:
    allergy = True

  if (prefer == True) and (dont_prefer == True):
    prefer = False
  if (dont_prefer == False):
    prefer = True
  if (non_vegetarian == True) and (vegetarian == True):
    vegetarian = False

  property_check = []

  if prefer == True:
    property_check.append('prefer')
  if dont_prefer == True:
    property_check.append('dont_prefer')
  if non_vegetarian == True:
    property_check.append('non_vegetarian')
  if vegetarian == True:
    property_check.append('vegetarian')
  if vegan == True:
    property_check.append('vegan')
  if diabetes == True:
    property_check.append('diabetes')
  if keto == True:
    property_check.append('keto')
  if allergy == True:
    property_check.append('allergy')

  food_display = []
  food_display_2 = []

  for key in food_property.keys():
    food_display.append(key)

  no_prefer_check = False

  for value in property_check:
    for food in food_display:
      if 'dont_prefer' in property_check:
        no_prefer_check = True
        if (value not in food_property[food]) and (value != 'dont_prefer') : #this still adds everything, so will need to check
          food_display_2.append(food)
          food_display_2 = list(set(food_display_2))
      else:
        if (value in food_property[food]):
          food_display_2.append(food)
          food_display_2 = list(set(food_display_2))

    filtered_food_display_2 = food_display_2[:]
    for food in food_display_2:
      for value in property_check:
        if 'dont_prefer' in property_check:
          if (value in food_property[food]) and (value != 'dont_prefer'):
             filtered_food_display_2.remove(food)
        else:
          if (value not in food_property[food]) and (value != 'prefer'):
             filtered_food_display_2.remove(food)



  if no_prefer_check == True:
    property_check.remove('dont_prefer')
    no_prefer_string = ' and '.join(property_check)
    print("We have detected that you don't prefer: " + no_prefer_string)
    print('Here are the item items that we think you will like:')
    for food in filtered_food_display_2:
      print(food.title())

    print('Would you like any of these items?')
    user_input = (input("You: ")).lower()

    if user_input == 'yes':
      return menu_options()
    elif user_input == 'no':
      print('Alright, thank you for visiting us!')
    else:
      print("Sorry, I didn't understand")

  else:
    property_check.remove('prefer')
    prefer_string = ' and '.join(property_check)
    print("We have detected that you prefer: " + prefer_string)
    print('Here are the item items that we think you will like:')
    for food in filtered_food_display_2:
      print(food.title())

    check = True
    while check == True:
      print('Would you like any of these items?')
      user_input = (input("You: ")).lower()
      confirmation_response = find_similarity(spell_check(user_input), yes_list, no_list, 'yes', 'no')

      if user_input == 'back':
        return food_recommendation()
      elif user_input == 'reset':
        return reset()

      if confirmation_response == 'yes':
        check = False
        food_recommendation_check = True
        return menu_options()
      elif confirmation_response == 'no':
        print('Alright, thank you for visiting us!')
        check = False
      else:
        print("Sorry, I didn't understand")
        continue

#=================================================================================================================================================================================================================

def issue_report():
    global report_issued
    issue_reported = False
    while True:
        if not issue_reported:
            print("Please describe the issue you're experiencing:")
            issue = input("Your issue: ")
            report_issued.append(issue)
            print("You reported:", issue)
            issue_reported = True

        print("Would you like to report another issue?")
        user_input = (input("You: ")).lower()
        confirmation_response = find_similarity(spell_check(user_input), yes_list, no_list, 'yes', 'no')

        if user_input == 'back':
          return issue_report()
        elif user_input == 'reset':
          return reset()

        if confirmation_response == 'yes':
            issue_reported = False  # Reset to allow reporting another issue
        elif confirmation_response == 'no':
            print("We apologize for the inconvenience and thank you for your feedback.")
            print("Returning you to the main selection page...")
            return main()
        else:
            print("Sorry, I din't understand.")
```

Secondary Path Functions

```python
def menu_options():
  global menu
  global starters_cost
  global main_course_cost
  global dessert_cost
  global beverages_cost
  global orders_food_check
  global food_recommendation_check

  check = True
  while check == True:
    print('Please type out your order now:')
    user_input = (input("You: ")).lower()
    menu_options_response = user_input

    if user_input == 'back':
      if orders_food_check == True:
        orders_food_check = False
        return orders_food()
      elif food_recommendation_check == True:
        food_recommendation_check = False
        return food_recommendation()
    elif user_input == 'reset':
      return reset()


    menu_doc = []
    menu_options_doc = nlp(menu_options_response)

    for item in menu['starters']:
      menu_doc.append(nlp(item))
    for item in menu['main course']:
      menu_doc.append(nlp(item))
    for item in menu['dessert']:
      menu_doc.append(nlp(item))
    for item in menu['beverages']:
      menu_doc.append(nlp(item))

    order_list = []
    order_quantity = []
    starters_list = []
    starters_quantity = []
    main_course_list = []
    main_course_quantity = []
    dessert_list = []
    dessert_quantity = []
    beverages_list = []
    beverages_quantity = []

    for token in menu_options_doc:
      for item in menu_doc:
        quantity_check = False

        if token.similarity(item) > 0.8:
          if item not in order_list:
            order_list.append(item)
          else:
            quantity_check = True

          for child in token.children:
            if child.dep_ == 'nummod':
              if quantity_check == True:
                order_quantity[-1] = child.text
              else:
                order_quantity.append(child.text)
              quantity_check = True

          if quantity_check == False:
            order_quantity.append(1)

    if not order_list:
      print('Sorry, but no menu items were detected.')
      return menu_options()

    for i, item in enumerate(order_list):
        if str(item) in menu['starters']:
            starters_list.append(item)
            starters_quantity.append(order_quantity[i])
        if str(item) in menu['main course']:
            main_course_list.append(item)
            main_course_quantity.append(order_quantity[i])
        if str(item) in menu['dessert']:
            dessert_list.append(item)
            dessert_quantity.append(order_quantity[i])
        if str(item) in menu['beverages']:
            beverages_list.append(item)
            beverages_quantity.append(order_quantity[i])

    total_cost = 0

    print('')
    for i, item in enumerate(starters_list):
      print('Starter ---------> ' + str(item) + ' X' + str(starters_quantity[i]))
      print('Cost ------------> $' + str(starters_cost[str(item)]*int(starters_quantity[i])))
      total_cost += starters_cost[str(item)]*int(starters_quantity[i])
      print('')
    for i, item in enumerate(main_course_list):
      print('Main Course -----> ' + str(item) + ' X' + str(main_course_quantity[i]))
      print('Cost ------------> $' + str(main_course_cost[str(item)]*int(main_course_quantity[i])))
      total_cost += main_course_cost[str(item)]*int(main_course_quantity[i])
      print('')
    for i, item in enumerate(dessert_list):
      print('Dessert ---------> ' + str(item) + ' X' + str(dessert_quantity[i]))
      print('Cost ------------> $' + str(dessert_cost[str(item)]*int(dessert_quantity[i])))
      total_cost += dessert_cost[str(item)]*int(dessert_quantity[i])
      print('')
    for i, item in enumerate(beverages_list):
      print('Beverage --------> ' + str(item) + ' X' + str(beverages_quantity[i]))
      print('Cost ------------> $' + str(beverages_cost[str(item)]*int(beverages_quantity[i])))
      total_cost += beverages_cost[str(item)]*int(beverages_quantity[i])
      print('')
    print('=============================================================================================================================================================================================')
    print('Total Cost ------> $' + str(total_cost))
    print('=============================================================================================================================================================================================')

    print("May I confirm this order?")
    user_input = (input("You: ")).lower()
    menu_response = find_similarity(spell_check(user_input), yes_list, no_list, 'yes', 'no')

    if user_input == 'back':
      check = False
      return menu_options()
    elif user_input == 'reset':
      check = False
      return reset()

    if menu_response == 'yes':
      print('Thank you for your confirmation')
      check = False
      return confirm_payment()
    elif menu_response == 'no':
      continue

#=================================================================================================================================================================================================================

def confirm_payment():
    print("Please confirm your order. Would you like to pay with Credit Card or Cash?")

    credit_card_list = ['card', 'credit', 'credit card']
    cash_list = ['money', 'physical notes', 'notes', 'cash', 'note']

    while True:  # Keep asking until a valid input is given
        user_input = input("You: ").lower()
        payment_response = find_similarity(spell_check(user_input), credit_card_list, cash_list, 'credit card', 'cash')

        if user_input == 'back':
          return menu_options()
        elif user_input == 'reset':
          return reset()

        if payment_response == 'credit card':
            return process_credit_card_payment()
        elif payment_response == 'cash':
            return process_cash_payment()
        else:
            print("Sorry, we only accept Credit Card, Cash, or Cashless payments. Please specify your payment method.")

    print("Thank you for your payment. Have a great day!")

def process_credit_card_payment():
  print("Please swipe your Credit Card at the terminal or insert it for chip payment.")
  # Simulate payment processing
  print("Processing your credit card payment...")
  return payment_acknowledged()

def process_cash_payment():
    print("Please proceed to the counter to pay with cash.")
    # Simulate payment processing
    print("Processing your cash payment...")
    return payment_acknowledged()

def payment_acknowledged():
    print("Your payment has been processed successfully.")
```
